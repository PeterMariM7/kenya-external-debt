```python
import requests
import pandas as pd
from sqlalchemy import create_engine
```

```python
country = 'KE'
indicator = 'DT.DOD.DECT.CD'
start_year = 2010
end_year = 2024


url = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}?date={start_year}:{end_year}&format=json&per_page=500"
response = requests.get(url)
response
```

```python
data = response.json()
records = data[1]
```

```python
df =pd.DataFrame(records)
df
```

```python
df['country_name'] = df['country'].apply(lambda x: x['value'])
```

```python
df_clean = df[['date', 'value','country_name']]
df_clean
```

```python
df_clean.columns = ['year', 'debt', 'country_name']
df_clean
```

```python
df_clean = df_clean.dropna(subset=['debt'])
```

```python
df_clean =df_clean.sort_values(by='year')
```
